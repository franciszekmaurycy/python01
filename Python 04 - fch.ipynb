65. wszystko na raz

```python
def discount(price, /, tax = 0.23, *, discount_rate = 0.1):
    """Oblicz cenę końcową po doliczeniu podatku i odjęciu rabatu.
    
    Args:
        price: Cena początkowa (tylko pozycyjnie).
        tax: Podatek VAT (domyślnie 0.23).
        discount_rate: Rabat (tylko keyword, domyślnie 0.1).
    
    Returns:
        Cena końcowa po zastosowaniu podatku i rabatu.
    """
    return ((price + (price * tax)) * (1 - discount_rate))

discount(100)                                # ➜ 100 * 1.23 * (1 - 0.1)
discount(100, tax=0.08)                      # ➜ 100 * 1.08 * (1 - 0.1)
discount(100, discount_rate=0.2)             # ➜ 100 * 1.23 * (1 - 0.2)
discount(100, tax=0.05, discount_rate=0.25)  # ➜ działa wszystko

"""price → tylko pozycyjnie,

tax → można pozycją albo keyword,

discount_rate → tylko keyword,"""
```

```python
def build_path(*folders: str, root: str = "/"):
    if not folders:  # jeśli nie podano żadnych folderów
        return root
    return root.rstrip("/") + "/" + "/".join(str(x) for x in folders if x is not None)

assert build_path("user", "docs", "images") == "/user/docs/images"
assert build_path("projects", root="/home/franek") == "/home/franek/projects"
assert build_path() == "/"

# bardziej zaawansowana technika, w połowie zrobiona przeze mnie
```

72. sortowańsko z lambdą

```python
cars = [
    {"brand": "A", "speed": 180, "price": 30000},
    {"brand": "B", "speed": 200, "price": 50000},
    {"brand": "C", "speed": 190, "price": 25000},
]

def top_n(items, scorer, n=3):
    return sorted(items, key=scorer, reverse=True)[:n]

# najtańsze
print(top_n(cars, lambda d: -d["price"]))

# najszybsze
print(top_n(cars, lambda d: d["speed"]))

# speed ↓, price ↑
print(top_n(cars, lambda d: (-d["speed"], d["price"])))
```

```python
licznik = 0  # GLOBAL

def outer():
    licznik = 0  # ENCLOSING (dla inner)
    
    def inner():
        nonlocal licznik # czyli enclosing, ale enclosing jest importowany tutaj do funkcji, pracujemy na tej funkcji z tą zmienną outer, jej używamy też jako inner. zatem zmieniamy dwie funkcje
        licznik += 1
        print(f"W środku (inner): {licznik}")

    inner()
    print(f"W zewnętrznej (outer): {licznik}")

outer()
print(f"Na samym dole (global): {licznik}")
```

79. Closure - domknięcia

```python
def stworz_skarbonke(kwota_startowa):
    
    def operacja(kwota):
        nonlocal kwota_startowa
        
        if kwota >= 0:
            kwota_startowa += kwota
            return kwota_startowa
        
        else:
            if kwota_startowa + kwota >= 0:
                kwota_startowa += kwota
                return kwota_startowa
            else:
                print("Brak wystarczających środków")
                return kwota_startowa
    
    return operacja

skarbonka = stworz_skarbonke(100)
print (skarbonka(1000))
print (skarbonka(-800))
print (skarbonka(-299))
print (skarbonka(9999))
print (skarbonka(-11111))

# closure to funkcja + środowisko zmiennych w którym powstało. zewnętrzna + wewnętrzna tworząca funkcję zewnętrzną, zmieniającą jej wartości. wewnętrzna funkcja zapamiętauje środowisko zewnętrznej funkcji dzięki nonlocal i zmiennej kwota_startowa
# funkcja niby się zamknęła, ale mamy ten "plecak" który pozwala nam na zachowanie wartości
# a to dzięki przekazaniu wartości przez nonlocal - pamiętaj, żeby używać na początku. tak jak import np. nie wewnątrz przy np. if
# normalnie w wewnętrznej funkcji kwota_startowa by nie istaniaa, ale ją wrzucamy do funkcji stwórz_skarbonke i wtedy dalej żyje. 
```

78. Opanowanie Kontraktów Funkcji i Walidacji Danych.

generalnie korzystamy z assert, żeby potweirdzić zgodność. ale to raczej wersja dla programisty do podglądu na bieżąco zgodności. 

w kodzie produkcyjnym w celu jego optymalizacji wykonuje się "raise" zamaist assert. jeśli coś wystąpi, wrzućmy raise "jakiś log"

no i można po raise krzyrzeć, typu "TypeError", "ValueError", "ZeroDivisionError". albo można zignorować sprawę. zauważyć ją podnosząc raise i to kończy nam działanie.

jeśli coś wiemy że będzie błęde, to dajemy return None lub pass i kontynuujemy działanie. raise kończy działanie.

jeśli czegoś szukamy i czegoś nie ma, to nie jest to błąd systemu tylko informacja o braku wyniku. możemy przechwycić błąd i uznać go na normalną, przewidywaną część naszego kodu. try except to poduszka powietrzna. 

```python
def calculate_calories(weight, duration, intensity):
    """ funkcja oblicza spalone kalorie podczas treningu
    argumenty:
    weight - waga użytkownika
    duration - czas trwaniatreningu w minutach
    intensity - jedno z low/medium/high, oznacza intensywność treningu
    
    returns:
    funkcja zwraca ilosć spalonych kalorii za pomocą wzoru mnożącego argumenty
    
    raises:
    typeerror lub valueerror jeśli wystąpi błąd dla danych
    
    przykłady użycia są na końcu xd
    """
    if not isinstance(weight, (int, float)):
        raise TypeError
    
    if not isinstance(duration, (int, float)):
        raise TypeError
    
    if weight <= 0 or duration <= 0:
        raise ValueError ("waga i czas nie mogą być ujemne")
    
    if not isinstance(intensity, str):
        raise ValueError ("dozwolone to low, medium lub high")
    
    multipliers = {
        "low": 0.05, 
        "medium": 0.08, 
        "high": 0.12
    }

    if intensity not in multipliers:
        raise ValueError ("hej, nie dozwolona wartość dla intensywności")
    
    value_intensity = multipliers[intensity]
    calories = weight * duration * value_intensity

    assert calories > 0, "Błąd w logice bracie"

    return calories

abc = calculate_calories(78, 3, "medium")
print(abc)

```

79. Pre i post warunki (kontrakt funkcji)

```python
def pierwiastek(liczba: float) -> float:
    # --- PRE-WARUNKI (Sprawdzamy na wejściu) ---
    if not isinstance(liczba, (int, float)):
        raise TypeError("Bracie, to musi być liczba!")
    if liczba < 0:
        raise ValueError("Nie liczmy pierwiastków z liczb ujemnych w tym programie.")

    # PRZED ROZPOCZĘCIEM DZIAŁANIA

    # ... tu dzieje się magia obliczeń ...
    wynik = liczba ** 0.5

    # --- POST-WARUNKI (Sprawdzamy wynik przed oddaniem) ---
    # Upewniamy się, że wynik ma sens (np. jest dodatni)
    assert wynik >= 0, "Ups, matematyka mi się popsuła"

    # PO ROZPOCZĘCIU DZIAŁANIA
    
    return wynik
```

82. praktyka pre, post warunki, eafp

```python
def bezpieczne_dzielenie(a,b):

    # pre
    if not isinstance(a, (int, float)):
        raise TypeError ("hej, a nie jest floatem")
    if not isinstance(b, (int, float)):
        raise TypeError ("hej, b nie jest floatem")
    
    # easier ask for permission than forgiveness
    try:
        wynik = a / b
    except ZeroDivisionError:
        return None
    # zakładam, że jest poprawne
    
    # look before you leap
    if b == 0:
        return None
    wynik = a / b
    # zakładam, że może nie być poprawne - bardziej pythonowo podobno

    # post
    assert isinstance(wynik, float), "Błąd, wynik nie jest floatem"

    return wynik

print(bezpieczne_dzielenie(23343215234534,1023452345))

def oblicz_rabat(cena, rabat):
    nowa_cena = cena * (1 - rabat)

    assert nowa_cena >= 0
    # sama asercja wewnątrz funkcji

    return nowa_cena

print(oblicz_rabat(100, 0.2))
```
