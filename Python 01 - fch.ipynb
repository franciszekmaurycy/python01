ZADANIE PODSUMOWUJĄCE LEKCJĘ 1

Cel: Funkcja `calculate(a, op, b)`, która liczy działanie matematyczne na dwóch liczbach `a` i `b`, gdzie `op` to jeden z operatorów:
`+ - * / // % **`

* `a` i `b` są liczbami (`int` lub `float`).

* `op` to tekst (np. `"+"`).

Zadbamy o **błędy** (np. dzielenie przez zero) i czytelne **komunikaty**.

Myśl o tym jak o małej „maszynce”: wrzucasz `a`, `op`, `b` → wychodzi wynik albo grzeczny komunikat o błędzie.

```python
a = 11
b = 7
# op = "+" or "-" or "*" or "**" or "/" or "//" or "%"
op = "+"

#assert isinstance(x, str), "bracie to nie jest string"
```

```python
def calculate (a, op, b):
    assert isinstance(a, int or float), "bracie, twój operand a to nie int ani float"
    assert isinstance(b, int or float), "bracie, twój operand b to nie int ani float"
    assert op == ("+", "-", "*", "**", "/", "//", "%")
    if b == 0:
        raise ZeroDivisionError("bracie, nie dzielimy przez zero")
    else:
        if op == "+":
            print (a + b)
        elif op == "-":
            print (a - b)
        elif op == "*":
            print (a * b)
        elif op == "**":
            print (a ** b)
        elif op == "/":
            print (a / b)
        elif op == "//":
            print (a // b)
        elif op == "%":
            print (a % b)
        else:
            print("bracie, nieprawidłowy operator!")
       
    
try:
    calculate
except AssertionError as xd:
    print (xd)
except ZeroDivisionError as lol:
    print (lol)
else:
    "zajebista robota bracie!"
finally: 
    "zawsze w Ciebie wierzyłem!"
```

```python
# Zestaw dozwolonych operatorów (łatwo tu sprawdzić, czy operator jest prawidłowy)
ALLOWED_OPERATORS = {"+", "-", "*", "/", "//", "%", "**"}

# Własny typ błędu, aby użytkownik dostawał czytelne komunikaty
class CalculatorError(Exception):
    """Błąd kalkulatora — np. nieznany operator albo dzielenie przez zero."""
    pass
```

```python
def calculate(a, op, b):
    """Zwraca wynik działania: a <op> b.

    a, b: liczby (int lub float)
    op: tekst, jeden z ALLOWED_OPERATORS
    """
    # 1) Sprawdź typy a i b
    # Jeżeli a lub b nie są liczbami (int/float), rzuć TypeError z czytelnym komunikatem.
    # TODO: dopisz warunek i podnieś TypeError
    # PRZYKŁAD komunikatu: "Argumenty muszą być liczbami (int lub float)."
    # --- TWÓJ KOD TUTAJ ---
    assert isinstance (a, int or float), CalculatorError
    assert isinstance (b, int or float), CalculatorError
    # 2) Sprawdź, czy operator jest dozwolony
    # Jeśli nie, rzuć CalculatorError z komunikatem "Nieznany operator: <op>"
    # --- TWÓJ KOD TUTAJ ---
    if op not in ALLOWED_OPERATORS:
        raise CalculatorError("not allowed")
    # 3) Specjalny przypadek: dzielenie przez zero
    # Dla operatorów "/", "//", "%" — jeśli b == 0 → CalculatorError("Dzielenie przez zero")
    # --- TWÓJ KOD TUTAJ ---
    if (op in ("/", "//", "%")) and (b == 0):
        raise CalculatorError("Dzielenie przez zero")
    # 4) Właściwe liczenie — zwróć wynik
    # Wypełnij bloki if/elif. Każdy ma zwrócić wynik działania.
    if op == "+":
        # TODO: zwróć a + b
        # --- TWÓJ KOD TUTAJ ---
        return (a + b)
    elif op == "-":
        # --- TWÓJ KOD TUTAJ ---
        return (a - b)
    elif op == "*":
        # --- TWÓJ KOD TUTAJ ---
        return (a * b)
    elif op == "/":
        # --- TWÓJ KOD TUTAJ ---
        return (a / b)
    elif op == "//":
        # --- TWÓJ KOD TUTAJ ---
        return (a // b)
    elif op == "%":
        # --- TWÓJ KOD TUTAJ ---
        return (a % b)
    elif op == "**":
        # --- TWÓJ KOD TUTAJ ---
        return (a ** b)
        # Jeśli tu trafimy, znaczy że coś pominęliśmy:
    else:
        raise CalculatorError("Nieobsłużony przypadek operatora.")
```

```python
# ====== ODKOMENTUJ, GDY UZUPEŁNISZ FUNKCJĘ 'calculate' ======
assert calculate(2, "+", 3) == 5
assert calculate(2, "-", 5) == -3
assert calculate(3, "*", 4) == 12
assert calculate(7, "/", 2) == 3.5
assert calculate(7, "//", 2) == 3
assert calculate(7, "%", 3) == 1
assert calculate(2, "**", 5) == 32
print("Działania podstawowe działają.")
```
