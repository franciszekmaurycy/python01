Najniższe koszty mają operacje stałe O(1) (np. dostęp do elementu listy, słownika lub zbioru), średnie — liniowe O(n) (np. przejście po liście, wyszukiwanie w niej, kopiowanie danych), a droższe — kwadratowe lub logarytmiczne O(n²), O(n log n) (np. podwójne pętle, sortowanie).
Aby utrzymać niskie koszty, warto używać struktur o szybkim dostępie, takich jak dict i set, zamiast list przy sprawdzaniu przynależności lub wyszukiwaniu.
Celem myślenia w Big-O jest unikanie zbędnych iteracji, kopiowań i sortowań — czyli pisanie kodu, który wykonuje jak najmniej pracy w stosunku do ilości danych.
