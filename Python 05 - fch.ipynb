91. json load i dump z with open. zapis i odczyt

```python
import json

studenci = [
    {"imie": "Anna", "wiek": 20, "ocena": 4.5},
    {"imie": "Jan", "wiek": 21, "ocena": 5.0},
    {"imie": "Ola", "wiek": 19, "ocena": 4.0}
]

# TODO: Zapisz listę do pliku "studenci.json"

with open("studenci.json", "w") as t:
    json.dump(studenci, t)

# TODO: Wczytaj i wypisz tylko studentów z oceną >= 4.5

with open("studenci.json", 'r') as t:
    dane = json.load(t)

# z Path musisz używać json.loads
# z with open używasz json.load

for student in dane:
    if student["ocena"] >= 4.5:
        print(student["imie"])
```

92. zapis i odczyt danych w json z wykorzystaniem Path i wyłapywanie błędów z try, except

```python
from pathlib import Path
import json

def zapisz_dane(dane, nazwa_pliku):
    """
    Zapisz dane do JSON. Zwróć True jeśli się udało, False jeśli błąd.
    """
    try:
        # TODO: zapisz dane do pliku
        Path(nazwa_pliku).write_text(json.dumps(dane))
        return True
    except Exception as e:
        print(f"Błąd zapisu: {e}")
        return False

def wczytaj_dane(nazwa_pliku):
    """
    Wczytaj JSON. Jeśli błąd, zwróć None.
    """
    try:
        # TODO: wczytaj dane z pliku
        return json.loads(Path(nazwa_pliku).read_text())
    except FileNotFoundError:
        return None
    except json.JSONDecodeError:
        return None

# Test:
dane = {"test": 123}
if zapisz_dane(dane, "test.json"):
    print("Zapisano!")

wczytane = wczytaj_dane("test.json")
print(wczytane)
```

94. odczyt i zmiana ustawień json w ramach klasy

```python
import json
from pathlib import Path

class Konfiguracja:
    def __init__(self, nazwa_pliku="config.json"):
        self.nazwa_pliku = nazwa_pliku
        self.ustawienia = self.wczytaj()
    
    def wczytaj(self):
        """Wczytaj konfigurację z pliku. Jeśli nie ma, zwróć domyślną."""
        try:
            # TODO: wczytaj z pliku
            return json.loads(Path(self.nazwa_pliku).read_text())
        except FileNotFoundError:
            # Zwróć domyślną konfigurację
            return {
                "jezyk": "pl",
                "motyw": "jasny",
                "rozmiar_czcionki": 12
            }
    
    def zapisz(self):
        """Zapisz konfigurację do pliku."""
        # TODO: zapisz self.ustawienia do pliku
        Path(self.nazwa_pliku).write_text(json.dumps(self.ustawienia))
    
    def ustaw(self, klucz, wartosc):
        """Zmień ustawienie i zapisz."""
        self.ustawienia[klucz] = wartosc
        self.zapisz()

# Test:
config = Konfiguracja()
print(config.ustawienia)
config.ustaw("motyw", "ciemny")
config.ustaw("rozmiar_czcionki", 14)
```
