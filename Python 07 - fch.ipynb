99. MRO

```python
class Jednostka:
    def __init__(self, imie):
        print(f"-> Inicjalizacja Jednostki: {imie}")
        self.imie = imie
    
    def atak(self):
        print("koniec atak")

# ta metoda musi być głucha i nie może mieć żadnego super

class Wojownik(Jednostka):
    def __init__(self, imie):
        print(f"-> Inicjalizacja Wojownika: {imie}")
        # TU BRAKUJE SUPER()
        super().__init__(imie)
        self.sila = 10
    
    def atak(self):
        print("atak")
        super().atak() # najpierw print, potem atak, jesli chcemy mieć print w dobrej kolejności
# generalnei możemy ustawić super przed lub po lub pomiędzy działaniem. jeśli działanie ma objąć dziedziczenie
# i np. przekazać informacje, to warto, żeby był ustawiony po wyzwloeniu informacji np.


class Uzdrowiciel(Jednostka):
    def __init__(self, imie):
        print(f"-> Inicjalizacja Uzdrowiciela: {imie}")
        # TU BRAKUJE SUPER()
        super().__init__(imie)
        self.mana = 20
    
    def atak(self):
        print("atak")
        super().atak()

# Paladyn dziedziczy po Wojowniku I Uzdrowicielu (kolejność ma znaczenie!)
class Paladyn(Wojownik, Uzdrowiciel):
    def __init__(self, imie):
        print(f"-> Tworzenie Paladyna: {imie}")
        # TU TEŻ UŻYJ SUPER(), żeby odpalić lawinę
        super().__init__(imie) # widzisz - super().metoda()

    def atak(self):
        print("atak")
        super().atak()

# --- TEST ---
# 1. Stwórz Paladyna
p = Paladyn("Arthas")

# 2. Sprawdź MRO
print("\nKolejność szukania (MRO):")
print(Paladyn.mro())

print(p.atak())

# super().metoda() - gdy mamy zwykłe self, automatycznie przekazywane jest wtedy
# super().metoda(arg) - gdy argument jest wymagany w łańcuchu przez następną klasę. jawnie
# super().metoda(*args,**kwargs) - w wielokrotnym dizeziczeniu by przepuścić nieużywane argumenty. rozpakowanie i przekazanie.
```

102. dataclass jeszcze raz

```python
class User:
    def __init__(self, name, age, email):
        self.name = name
        self.age = age
        self.email = email

    def __repr__(self):
        return f"User(name={self.name}, age={self.age}, email={self.email})"

    def __eq__(self, other):
        if not isinstance(other, User): return False
        return self.name == other.name and self.age == other.age and self.email == other.email

    # TAK TO WYGLĄDA BEZ DATACLASS

u = User("franek", 21, "python@onet.pl")
print(u)
```

```python
from dataclasses import dataclass

@dataclass
class User:
    name: str
    age: int
    email: str

u = User("franek", 21, "python@onet.pl")
print(u)

# TAK TO WYGLĄDA Z DATACLASS
```
